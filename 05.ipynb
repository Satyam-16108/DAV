# Taking Iris data , plot the following with proper legend and axis labels: ( Download IRIS the data from : https://achive.ics.edu/ml/datasets/iris or import it from sklearn.datasets)

```python
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris_data=load_iris()
```

```python
iris_data.keys()
```

```python
df=pd.DataFrame(iris_data['data'],columns=['sepal_length',
 'sepal_width',
 'petal_length',
 'petal_width'])
df['target']=iris_data['target']
df
```

```python
iris_data['target']
```

# a. Plot bar chart to show the frequency of each class label in the data. 

```python
setosa_freq=list(df.iloc[:,4]).count(0)
versicolour_freq=list(df.iloc[:,4]).count(1)
virginica_freq=list(df.iloc[:,4]).count(2)
names=list(['setosa','versicolour','virginica'])
freq=list([setosa_freq,versicolour_freq,virginica_freq])
```

```python
plt.bar(names,freq,color=['Red','Blue','Green'], width=0.5)
plt.xlabel('Iris type');
plt.ylabel('Frequency')
plt.show()
```

# b.Draw the scatter plot for petal width vs sepal width

# c. Plot density distribution for feature petal width

```python
df.petal_width.plot.density(color='green')
plt.title("Density distribution for Petal Width")
plt.show()
```
