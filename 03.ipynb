Create a dataframe having at least 3 columns and 50 rows to store numeric data generated using a random function.
Replace 10% of the values by null values whose index positions are generated using random function.

```python
import pandas as pd
import numpy as np
arr=np.random.randint(10,size=(50,3))
n=15
df=pd.DataFrame(arr)
while(n):
    y=np.random.randint(1,50)
    x=np.random.randint(0,3)
    df.loc[y,x]=np.nan
    n=n-1
print(df)
```

# a. Identify and count missing values in dataframe.

```python
count=0
a=df.isnull()
df1=pd.DataFrame(a)
arr=[]
for i in range(50):
    for j in range(3):
        if(df1.loc[i,j]):
            count=count+1
            arr.append([i,j])
print('The missing values are :',count)
print('At the indices:',arr)
```

# b. Drop the column having more than 5 null values

```python
for col in df.columns:
    print(col,df[col].isnull().sum())
df.dropna(axis=1,thresh=(df.shape[0]-5)).head()
```

# c. Identify the row label having maximum of the sum of all the values in a row and drop the row

```python
sum=df.sum(axis=1)
print("Sum is :\n",sum)
print("\n Maximum sum is: ", sum.max())
max_sum_row=df.sum(axis=1).idxmax()
print("\n Row index having max sum is: ",max_sum_row)
df=df.drop(max_sum_row,axis=0)
print("\n dataframe after removing the row having max sum value")
df
```

# e. Remove all duplicates from the first column

```python
df=df.drop_duplicates(subset=0,keep="first")
print(df)
```

# f.  Find the correlation between first and second column an dcovariance between second and third column

```python
df1=df.dropna()
corr = np.corrcoef(df1.iloc[:,0], df1.iloc[:,1] )
print(corr)

covv = np.cov(df1.iloc[:,1], df1.iloc[:,2])

print("covarience--->",covv)
```

# g. Detect the outliers and remove the rows having outliers

```python
import matplotlib.pyplot as plt
#data=df1.iloc[:,0]
data_arr  = np.array([-900,1,2,3,4])

q1 ,q2, q3  = np.percentile(sorted(data_arr) , [25,50,75])
print(sorted(data_arr))
print("total -->",data_arr.size)
print("q1 --> ",q1)
print("q2 --> ",q2)
print("q3 --> ",q3)

iqr = q3 - q1

print("iqr  --> ",iqr)

lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)

print("lower bound",lower_bound)
print("upper_bound",upper_bound)
outliers = [i for i in data_arr if i<= lower_bound or i >=upper_bound]
print(outliers)

plt.boxplot(data_arr)
plt.show()
```
